# KNN -- K NEAREST NEIGHBOURS
1. TP = True Positive
2. TN = True Negative
3. FP = FALSE POSITIVE
4. FN = FALSE Negative
if prediction is positive and actual was true so TP.

<!-- table -->
       |   predicted
       ----------------------
            |  no     |   yes
---------------------------
actual | no | tn      |   fp
       | yes| fn      |   tp
-----------------------------   

<!--accuracy -- HOW MANY PREDICTIONS ARE TRUE -->
=(tp+tn)/total

<!--error rate  -- HOW MANY PREDICTIONS ARE FALSE-->
=(fp+fn)/total ==1- accuracy

<!--sensitivity = positive direction ma model nu sachu padvano rate -->
=(tp)/tp+fn

<!-- specificity = NEGATIVE direction ma model nu SACHU padvano rate-->
=(tn)/tn+fp


```python
import pandas as pd
df=pd.read_csv('diabetes.csv')
df
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
y=df['Outcome']
x=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=25)
```

```python
model=nn.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
print('TN =',TN)
print('FP =',FP)
print('FN =',FN)
print('TP =',TP)
```

```python
TOTAL=TP+TN+FP+FN
```

```python
accuracy=(TP+TN)/TOTAL
accuracy
```

```python
err=(FP+FN)/TOTAL
err
```

```python
sensi=(TP)/(TP+FN)
sensi
```

```python
speci=(TN)/(TN+FP)
speci
```

```python
import pandas as pd
df=pd.read_csv('PlayTennis.csv')
df
```

```python
df=pd.get_dummies(data=df,drop_first=True)
df
```

```python
df.columns
```

```python
y=df['play_yes']
x=df[['windy', 'outlook_rainy', 'outlook_sunny', 'temp_hot', 'temp_mild',
       'humidity_normal']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=3)
# here 11 row in x_train so neighbors are = root of 11
```

```python
model=nn.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
print('TN =',TN)
print('FP =',FP)
print('FN =',FN)
print('TP =',TP)
```

```python
TOTAL=TP+TN+FP+FN
print("Accuracy =",(TP+TN)/TOTAL)
print("Error_Rate =",(FP+FN)/TOTAL)
print("Sensitivity =",(TP)/(TP+FN))
print("Specificity =",(TN)/(TN+FP))
```
