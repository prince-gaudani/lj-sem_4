**Current weather data**

```python
website="https://api.openweathermap.org/data/2.5/weather?lat="+str(Lat)+"&lon="+str(Long)+"&units=metric&appid="+api_key
result=requests.get(website)
weather_data=result.json()
print(json.dumps(weather_data,indent=5))
```

```python
description=weather_data['weather'][0]['description']
visibility=weather_data['visibility']
temp=weather_data['main']['temp']
pressure=weather_data['main']['pressure']
humidity=weather_data['main']['humidity']
wind_speed=weather_data['wind']['speed']
country=weather_data['sys']['country']
print("description=",description)
print("visibility=",visibility)
print("temp=",temp)
print("pressure=",pressure)
print("humidity=",humidity)
print("wind_speed=",wind_speed)
print("country=",country)
```

**Air Pollution**

```python
website="http://api.openweathermap.org/data/2.5/air_pollution?lat="+str(Lat)+"&lon="+str(Long)+"&appid="+api_key
result=requests.get(website)
Air_pollution=result.json()
print(json.dumps(Air_pollution,indent=5))
```

```python
aqi=Air_pollution['list'][0]['main']['aqi']
quality_indices={1:"Good", 2 : "Fair", 3 : 'Moderate', 4 : 'Poor', 5 : 'Very Poor'}
print("current air quality=",quality_indices[aqi])
```

**5 Day 3 Hour forecast**

```python
website="http://api.openweathermap.org/data/2.5/forecast?lat="+str(Lat)+"&lon="+str(Long)+"&appid="+api_key
result=requests.get(website)
five_day=result.json()
print(json.dumps(five_day,indent=5))
```

```python
date_time=[]
temp=[]
pressure=[]
humidity=[]
description=[]
for i in five_day['list']:
    date_time.append(i['dt_txt'])
    temp.append(i['main']['temp'])
    pressure.append(i['main']['pressure'])
    humidity.append(i['main']['humidity'])
    description.append(i['weather'][0]['description'])
data={'date_time':date_time,'temp':temp,'pressure':pressure,'humidity':humidity,'description':description}
import pandas as pd
df=pd.DataFrame(data)
df
```

**send query parameter**

**send custom headers**

```python
headers={'user-Agent':'Mozilla/5.0','Accept':'application/json'}
response=requests.get(url,headers=headers)
print(response.status_code)
```

**send POST requests**

```python
url='https://jsonplaceholder.typicode.com/posts'
new_post={'title':'Learning requets','body':"this is tutorial content",'userId':1}
response=requests.post(url,json=new_post)
print(response.status_code)
print(response.json())
```

**send PUT requests**

```python
url='https://jsonplaceholder.typicode.com/posts/1'
updated_post={'id':1,'title':'Updated title','body':'Updated body','userId':1}
response=requests.put(url,json=updated_post)
print(response.status_code)
print(response.json())
```

**send PATCH requsts**

```python
url='https://jsonplaceholder.typicode.com/posts/1'
patch_data={'title':"Patched Title"}
response=requests.patch(url,json=patch_data)
print(response.status_code)
print(response.json())
```

**send DELETE request**

```python
url='https://jsonplaceholder.typicode.com/posts/1'
response=requests.delete(url)
print(response.status_code)
```

**saved json data to file**

```python
url='https://jsonplaceholder.typicode.com/posts/1'
response=requests.get(url)
data=response.json()
with open('post1.json','w') as f:
    json.dump(data,f,indent=4)
```

**read json data from file**

```python
with open('post1.json','r') as f:
    loaded_data=json.load(f)
print(loaded_data)
```
