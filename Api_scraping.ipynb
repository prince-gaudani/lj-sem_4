# Current Weather

```python
# current whether data api from open wheather
# https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API key}
website='https://api.openweathermap.org/data/2.5/weather?lat='+str(lat)+'&lon='+str(lon)+'&units=metric&appid='+api_key
# str because lat and lon are float so convert into string
result=requests.get(website)
weather=result.json()
print(json.dumps(weather,indent=5))
# for proper print and indent for space
# here units=metric for temp in celcius
```

```python
dis=weather['weather'][0]['description']
vis=weather['visibility']
temp=weather['main']['temp']
hum=weather['main']['humidity']
windspeed=weather['wind']['speed']
press=weather['main']['pressure']
ccode=weather['sys']['country']
print('discription:',dis)
print('visiblity:',vis)
print('temp:',temp)
print('humidity:',hum)
print('Wind Speed:',windspeed)
print('Country Code:',ccode)
print('Pressure:',press)
```

# 5 day/3 hour forecast

```python
# give datframe of this data in from 5 day /3 hour forecast :date and time,temprature,pressure,humidity,discription
website='http://api.openweathermap.org/data/2.5/forecast?lat='+str(lat)+'&lon='+str(lon)+'&appid='+api_key
result=requests.get(website)
forecast=result.json()
print(json.dumps(forecast,indent=1))
```

```python
import pandas as pd 
datetime=[]
temp=[]
pressure=[]
humidity=[]
description=[]
for i in forecast['list']:
    temp1=i['main']['temp']
    hum=i['main']['humidity']
    press=i['main']['pressure']
    des=i['weather'][0]['description']
    dt=i['dt_txt']
    datetime.append(dt)
    temp.append(temp1)
    humidity.append(hum)
    pressure.append(press)
    description.append(des)
data={"Date Time":datetime,"Temprature":temp,"Humidity":humidity,"Pressure":pressure,"Discription":description}
df=pd.DataFrame(data)
```
