```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data={"Math":[80,70,60],"Science":[75,65,55],"English":[90,85,70]}
df=pd.DataFrame(data)
sns.heatmap(df)
plt.title("Students Marks")
plt.show()
```

# pb - 139
The following dictionary shows how five people follow each other on Instagram:
 instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
 E.g., the list for person1 has the value on index 2 as 1 which means person1 followsperson3 and a directed edge should be 
added from person1 to person3.
 
 Using networkx library, create a directed graph          <!--  here persion 1 follows person 2,3,5 because of 1-->


```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}

Node=list(instagram.keys())
G.add_nodes_from(Node) 

Follow=list(instagram.values())
for i in range(len(Follow)):
    for j in range(len(Follow)):
        if Follow[i][j]==1:
            G.add_edge(Node[i],Node[j])
            
nx.draw_networkx(G,node_size=2200,node_color="Orange",width=3,edge_color="Red")
plt.show()
```
