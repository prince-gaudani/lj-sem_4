# Finding and Removing Outliers

```python
# outliers - 

#  ex: in 1,20,21,19,49 series --> 1 and 49 are outliers
```

```python
df=pd.read_csv("auto-mpg.csv")
df
```

```python
Q1=df["acceleration"].quantile(0.25)
Q1
```

```python
Q3=df["acceleration"].quantile(0.75)
Q3
```

```python
Q1=df["acceleration"].quantile(0.25)
Q3=df["acceleration"].quantile(0.75)
IQR=Q3-Q1
LOW=Q1-(1.5*IQR)
HIGH=Q3+(1.5*IQR)
```

```python
print("LOW :",LOW)
print("High :",HIGH)
```

```python
Find_Outliers=df[(df["acceleration"]<LOW)|(df["acceleration"]>HIGH)]
```

```python
Find_Outliers
```

```python
Remove_Outliers=df[(df["acceleration"]>=LOW)&(df["acceleration"]<=HIGH)]
Remove_Outliers
```

# merge Dataframe

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2,3],"name":["A","B","C"],"dept_id":[10,20,30]})
departments=pd.DataFrame({"dept_id":[10,20,30],"dept_name":['HR',"IT","Finance"]})
pd.merge(employee,departments,on="dept_id",how='inner')
# here by default innner used without how
# here merging process according to columns
```

```python
employee=pd.DataFrame({"emp_id":[1,2,3],"name":["A","B","C"],"dept_id":[10,20,30]})
departments=pd.DataFrame({"dept_id":[10,20,40],"dept_name":['HR',"IT","Finance"]})
pd.merge(employee,departments,on="dept_id",how='outer')
```

```python
employee=pd.DataFrame({"emp_id":[1,2,3],"name":["A","B","C"],"dept_id":[10,20,30]})
departments=pd.DataFrame({"dept_id":[10,20,40],"dept_name":['HR',"IT","Finance"]})
pd.merge(employee,departments,on="dept_id",how='outer',indicator=True)
```

```python
employee=pd.DataFrame({"emp_id":[1,2,3],"name":["A","B","C"],"dept_id":[10,20,30]})
departments=pd.DataFrame({"dept_id":[10,20,40],"dept_name":['HR',"IT","Finance"]})
pd.merge(employee,departments,on="dept_id",how='left')
# it takes all data fram  left df and only matching data from right df
# here employee is left df
```

```python
employee=pd.DataFrame({"emp_id":[1,2,3],"name":["A","B","C"],"dept_id":[10,20,30]})
departments=pd.DataFrame({"dept_id":[10,20,40],"dept_name":['HR',"IT","Finance"]})
pd.merge(employee,departments,on="dept_id",how='right')
```

# left_on , right_on

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2,3],"name":["A","B","C"],"dept_id":[10,20,30]})
departments=pd.DataFrame({"d_id":[10,20,40],"dept_name":['HR',"IT","Finance"]})
pd.merge(employee,departments,left_on="dept_id",right_on="d_id")
# here exact common column title not same 
```

# merge using index(left_index,right_index)

```python
employee=pd.DataFrame({"emp_id":[1,2],"name":["A","B"],"dept_id":[10,20]})
departments=pd.DataFrame({"d_id":[10,20],"dept_name":['HR',"IT"]})
pd.merge(employee,departments,left_index=True,right_index=True)
# row wise both table merge
```

# suffixes

```python
employee=pd.DataFrame({"emp_id":[1,2],"name":["A","B"],"dept_id":[10,20]})
departments=pd.DataFrame({"dept_id":[10,20],"dept_name":['HR',"IT"],"name":["X","Y"]})
pd.merge(employee,departments,on="dept_id")
```

```python
employee=pd.DataFrame({"emp_id":[1,2],"name":["A","B"],"dept_id":[10,20]})
departments=pd.DataFrame({"dept_id":[10,20],"dept_name":['HR',"IT"],"name":["X","Y"]})
pd.merge(employee,departments,on="dept_id",suffixes=("_emp","_dept"))
```
