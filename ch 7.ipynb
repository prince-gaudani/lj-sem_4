```python
import requests 
from bs4 import BeautifulSoup
```

```python
website='https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
result=requests.get(website)
content=result.text
soup=BeautifulSoup(content,'html.parser')
```

```python
import pandas as pd
laptops=soup.find_all('div',class_='card thumbnail')
laptop_names=[]
laptop_price=[]
for i in laptops:
    laptop_names.append(i.find('a').text.strip())
    laptop_price.append(i.find('h4').text.strip())
data={'Laptop_names':laptop_names,'Laptop_price':laptop_price}
df=pd.DataFrame(data)
df.to_csv("laptop.csv",index=False)
```

```python
# for html web page (30 populer movies)
import requests 
from bs4 import BeautifulSoup
import pandas as pd
with open('30 Most Popular Movies.html','r',encoding='utf-8') as file:
    content=file.read()
soup=BeautifulSoup(content,'html.parser')
```

```python
movies=soup.find_all('div',class_='article_movie_title')
movie_name=[]
movie_rating=[]
for i in movies:
    movie_name.append(i.find('a').text.strip())
    movie_rating.append(i.find('span',class_='tMeterScore').text.strip())
data={"Movie_names":movie_name,'Movie_ratings':movie_rating}
df=pd.DataFrame(data)
df
```

```python
# fetch name,milege,price,rating from mercideze benze html file
import requests 
from bs4 import BeautifulSoup
import pandas as pd
with open('Certified used Mercedes-Benz for sale.html','r',encoding='utf-8') as file:
    content=file.read()
soup=BeautifulSoup(content,'html.parser')
```

```python
box=soup.find_all('div',class_='vehicle-details')
name=[]
mileage=[]
price=[]
rating=[]
for i in box:
    a=i.find('h2',class_='title')
    name.append(a.text)
    b=i.find('div',class_='mileage')
    mileage.append(b.text)
    c=i.find('span',class_='primary-price')
    price.append(c.text)
    d=i.find('span',class_='sds-rating__count')
    if d:
        rating.append(d.text.strip())
    else:
        rating.append(None)
data={"Car Name":name,"Car Mileage":mileage,"Car Price":price,"Car Rating":rating}
df=pd.DataFrame(data)
df
```

```python
# name,rating,discount,actual price
import requests 
from bs4 import BeautifulSoup
import pandas as pd
with open('flipkart.html','r',encoding='utf-8') as file:
    content=file.read()
soup=BeautifulSoup(content,'html.parser')
```

```python
box=soup.find_all('div',class_='_3pLy-c row')
name=[]
d_price=[]
a_price=[]
rating=[]
for i in box:
    a=i.find('div',class_='_4rR01T')
    name.append(a.text)
    b=i.find('div',class_='_30jeq3 _1_WHN1')
    d_price.append(b.text)
    c=i.find('div',class_='_3I9_wc _27UcVY')
    # for missing actual price
    if c:
         a_price.append(c.text.strip())
    else :
        a_price.append(None)
   
    d=i.find('div',class_='_3LWZlK')
    rating.append(d.text)

data={"Mobile Name":name,"Rating":rating,"Discount Price":d_price,"Actual Price":a_price}
df=pd.DataFrame(data)
df
```
